```python
import pandas as pd
```

```python
nodes = pd.read_csv('nodes.csv')
edges = pd.read_csv('edges.csv')
```

```python
nodes_disease = nodes[nodes['0']=='disease']
nodes_genes = nodes[nodes['0']=='gene']
```

```python
Names = {nodes.iloc[i]['Id']:nodes.iloc[i]['Label'] for i in range(len(nodes))}
Class = {Names[nodes.iloc[i]['Id']]:nodes.iloc[i]['1'] for i in range(len(nodes))}
```

```python
nodes_genes
```

```python
G = []
D = []
for i in range(len(edges)):
    if edges['Source'][i] in nodes_disease['Id'].unique() and edges['Target'][i] in nodes_disease['Id'].unique():
        D.append([Names[edges['Source'][i]],Names[edges['Target'][i]]])
    else:
        G.append([Names[edges['Source'][i]],Names[edges['Target'][i]]])
D = pd.DataFrame(D,columns=['Source','Target'])
G = pd.DataFrame(G,columns=['Source','Target'])
```

```python
G
```

```python
# Mapping between diseases and genes associated with them
Dictionary_D = {}
for i in G['Source'].unique():
    Dictionary_D[i]=list(G[G['Source']==i]['Target'].unique())
Dictionary_D
```

```python
# Mapping between genes and diseases associated with them
Dictionary_G = {}
for i in G['Target'].unique():
    Dictionary_G[i]=set(G[G['Target']==i]['Source'].unique())
Dictionary_G
```

# Node Size

```python
Size_D = {i:len(Dictionary_D[i]) for i in list(Dictionary_D.keys())}
Size_G = {i:len(Dictionary_G[i]) for i in list(Dictionary_G.keys())}
```

```python
Size_D
```

```python
L1 = list(zip(list(Size_D.keys()),list(Size_D.values())))
L2 = list(zip(list(Size_G.keys()),list(Size_G.values())))
```

```python
pd.DataFrame(L1,columns=['Gene','Size']).to_csv('Disease_size.csv')
pd.DataFrame(L2,columns=['Gene','Size']).to_csv('Gene_size.csv')
```

# Edge Size

```python
S = []
for i in range(len(D)):
    S.append(len(set(Dictionary_D[D.iloc[i]['Source']]) & set(Dictionary_D[D.iloc[i]['Target']])))
D['Common Genes'] = S
```

```python
D.to_csv('Disease_edges.csv')
```

# Genes Table

```python
Dictionary_G
```

```python
I = []
for i in Dictionary_G:
    for j in Dictionary_G:
        if i != j and len(Dictionary_G[i] & Dictionary_G[j])>0:
            I.append([i,j,len(Dictionary_G[i] & Dictionary_G[j])])
I = pd.DataFrame(I,columns=['Souce','Target','Common Diseases'])
```

```python
I.to_csv('Genes_edges.csv')
```

```python
for i in range(len(nodes_genes)):
    As_D = Dictionary_G[nodes_genes.iloc[i]['Label']]
    c = {Class[j] for j in As_D}
    if len(c) == 1:
        nodes_genes['1'][i] = list(c)[0]
    elif len(c) > 1:
        nodes_genes['1'][i] = 'Unclassified'
```
