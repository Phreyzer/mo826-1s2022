```python
cd1 = condicoes1[(condicoes1['DESCRIPTION'] == 'COVID-19') | (condicoes1['DESCRIPTION'] == 'Suspected COVID-19')]
cd2 = condicoes2[(condicoes2['DESCRIPTION'] == 'COVID-19') | (condicoes2['DESCRIPTION'] == 'Suspected COVID-19')]
cd3 = condicoes3[(condicoes3['DESCRIPTION'] == 'COVID-19') | (condicoes3['DESCRIPTION'] == 'Suspected COVID-19')]
cd4 = condicoes4[(condicoes4['DESCRIPTION'] == 'COVID-19') | (condicoes4['DESCRIPTION'] == 'Suspected COVID-19')]
```

```python
fp1 = pacientes1[['Id','BIRTHDATE','DEATHDATE','RACE','ETHNICITY','GENDER']]
fp1=fp1.rename(columns={'Id':'PATIENT'})

fp2 = pacientes2[['Id','BIRTHDATE','DEATHDATE','RACE','ETHNICITY','GENDER']]
fp2=fp2.rename(columns={'Id':'PATIENT'})

fp3 = pacientes3[['Id','BIRTHDATE','DEATHDATE','RACE','ETHNICITY','GENDER']]
fp3=fp3.rename(columns={'Id':'PATIENT'})

fp4 = pacientes4[['Id','BIRTHDATE','DEATHDATE','RACE','ETHNICITY','GENDER']]
fp4=fp4.rename(columns={'Id':'PATIENT'})
```

```python
inter1 = pd.merge(fp1,cd1,how='inner',on=['PATIENT'])
inter1.drop_duplicates(['PATIENT'],keep='first',inplace=True)
inter1.drop(['STOP','ENCOUNTER','CODE','DESCRIPTION'],axis=1,inplace = True)

inter2 = pd.merge(fp2,cd2,how='inner',on=['PATIENT'])
inter2.drop_duplicates(['PATIENT'],keep='first',inplace=True)
inter2.drop(['STOP','ENCOUNTER','CODE','DESCRIPTION'],axis=1,inplace = True)

inter3 = pd.merge(fp3,cd3,how='inner',on=['PATIENT'])
inter3.drop_duplicates(['PATIENT'],keep='first',inplace=True)
inter3.drop(['STOP','ENCOUNTER','CODE','DESCRIPTION'],axis=1,inplace = True)

inter4 = pd.merge(fp4,cd4,how='inner',on=['PATIENT'])
inter4.drop_duplicates(['PATIENT'],keep='first',inplace=True)
inter4.drop(['STOP','ENCOUNTER','CODE','DESCRIPTION'],axis=1,inplace = True)
```
